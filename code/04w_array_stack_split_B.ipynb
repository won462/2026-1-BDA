### numpy 함수 vstack hstack row_stack column_stack 비교 설명

| 항목 | `np.vstack` | `np.hstack` | `np.row_stack` | `np.column_stack` |
|---|---|---|---|---|
| **별칭 여부** | — | — | `vstack`의 alias | — |
| **연결 축** | axis=0 (수직) | axis=1 (수평) | axis=0 (수직) | axis=1 (수평) |
| **내부 구현** | `concatenate(axis=0)` + `atleast_2d` | 1D: `concatenate(axis=0)` / 2D+: `concatenate(axis=1)` | `vstack`과 동일 | 1D를 `(N,1)`로 변환 후 `concatenate(axis=1)` |
| **shape 요구 조건** | 열 수 일치 | 행 수 일치 | 열 수 일치 | 행 수 일치 |
| **Deprecated** | — | — | NumPy 2.0+ | — |

### np.random.random()	np.random.default_rng().random() 비교

| 항목    | np.random.random() | default_rng().random() |
| ----- | ------------------ | ---------------------- |
| 방식    | 전역 상태 사용           | 객체 기반                  |
| 알고리즘  | Mersenne Twister   | PCG64                  |
| 재현성   | 취약 (전역 영향 받음)      | 매우 안정적                 |
| 병렬 처리 | 부적합                | 적합                     |
| 권장 여부 | ❌ (구식)             | ✅ (권장)                 |


np.random.random() np.random.default_rng().random() 비교

| 항목 | `np.random.random()` (Legacy / RandomState) | `default_rng().random()` (New / Generator, NumPy 1.17+) |
|---|---|---|
| **내부 알고리즘** | Mersenne Twister (MT19937) | PCG64 (기본값); SFC64, MT19937 등 선택 가능 |
| **상태 크기** | 624 × 32-bit words (~2.5 KB) | PCG64: 128-bit state (16 B) |
| **주기 (period)** | 2^19937 − 1 | PCG64: 2^128 |
| **BigCrush 통과** | 일부 실패 (LinearComp, MatrixRank 등) | 통과 |
| **고차원 균일성** | MT는 623차원에서 균일, 그 이상 편향 가능 | PCG64: 고차원에서도 우수 |
| **시드 설정** | `np.random.seed(42)` — 전역 상태 변경 (side-effect) | `rng = np.random.default_rng(42)` — 로컬 객체, 독립적 |
| **전역 상태 오염** | 있음 — 다른 코드와 간섭 위험 | 없음 — 인스턴스 독립 |
| **버전 간 재현성** | NumPy 버전 간 보장 안 됨 | 동일 BitGenerator 사용 시 보장 |
| **독립 스트림 생성** | 불편 — 수동 시드 관리 필요 | 공식 지원 (`rng.spawn(n)` 또는 `SeedSequence`) |
| **속도** | 보통 | PCG64 > MT19937 일반적; SFC64는 더 빠름 |
| **추가 분포 API** | 기본 분포 제공 | `rng.integers()`, `rng.standard_normal()` 등 확장 API |
| **BitGenerator 교체** | 불가 | 가능 — `np.random.Generator(MT19937(42))` |
| **신규 코드 권장** | 비권장 | 권장 (NEP 19) |
| **레거시 호환** | 적합 | 일부 함수명 상이 (하위 호환 주의) |
| **ML/시뮬레이션 연구** | 재현성 관리 복잡 | 권장 — 실험 단위 RNG 분리 용이 |
